```python
import pandas as pd
import numpy as np
```

## Read in data

```python
full_data = pd.read_csv('../englishdata.csv', lineterminator="\n")
```

```python
full_data
```

```python
char_per_comp = full_data.groupby('Company_Name').sum('characters').sort_values('characters')
filtered_comps = char_per_comp[(char_per_comp['characters']<20000) & (char_per_comp['characters']>1000) & ~(char_per_comp['characters'].isna())]
filtered_comps
```

```python
companies = filtered_comps.sample(150, random_state=5001).index
```

```python
filtered_data = full_data[full_data['Company_Name'].isin(companies)]
len(filtered_data)
```

```python
filtered_data.groupby('Company_Name').first()
```

```python
filtered_data.to_csv('subset_data.csv', index=False)
```
