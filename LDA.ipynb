```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
VOCAB_LDA = pd.read_csv('./data/VOCAB_LDA.csv', index_col='term_str')
DOCS_LDA = pd.read_csv('./data/DOCS_LDA.csv', index_col='company_id')
TOPICS = pd.read_csv('./data/TOPICS.csv', index_col='topic_id')
THETA = pd.read_csv('./data/THETA.csv', index_col='company_id')
```

```python
TOPICS
```

```python
t = ['Manufacturing', 'Customer Rights', 'People', 'General Company Information', 'Product Parts',
 'Casting Process', 'Magnet Manufacturing', 'Long-Term Quality', 'Advanced Technology', 'Services', 
 'Industry', 'Production Techniques', 'Customer Privacy', 'Product Design/Team', 'Parts', 
 'Bearing', 'Glass', 'Website Information', 'Customer Service', 'Metal']
TOPICS['Overall_Topic'] = t
```

```python
TOPICS
```

```python
THETA_max = THETA.idxmax(axis=1).to_frame().rename(columns={0:'topic_id'})
```

```python
company_topics = TOPICS[['Overall_Topic']].merge(THETA_max, left_index=True, right_on='topic_id')
```

```python
company_topics.head()
```

```python
company_topics.to_csv('./visualization_data/company_topics.csv')
```

```python
ax = sns.histplot(data = company_topics, y='Overall_Topic')
ax.set(ylabel='Topic');
```
