# Visualize Loadings

```python
X = LOADINGS.join(VOCAB)
X.to_csv('./visualization_data/X.csv') #index=term_str
```

```python
def vis_loadings(a=0, b=1, hover_name='term_str'):
    X = LOADINGS.join(VOCAB)
    return px.scatter(X.reset_index(), f"PC{a}", f"PC{b}", 
                      text='term_str', size='i', color='max_pos',
                      marginal_x='box', height=800)

def vis_pcs(M, a, b, label='company_id', hover_name='company_id',symbol=None, size=None):
    fig = px.scatter(M, f"PC{a}", f"PC{b}", hover_name=hover_name, 
                     symbol=symbol, size=size,
                     marginal_x='box', height=800)
    fig.show()
```

```python
vis_pcs(DCM.reset_index(), 0, 1)
```

Companies that are high on PC0 have a good website/privacy policy and likely have a large social media or online website presence. Companies that are high on PC1 have a focus on their website on specific products they have and on the raw materials that go into those products.

```python
vis_loadings(0, 1)
```
